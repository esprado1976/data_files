```python
#Abstract
# El presente trabajo tomará el data set WeatherAUS.csv con datos de clima en australia con el fin de crear un modelo predictivo de probabiblidad
# de precipitaciones a partir de condiciones meteorológicas previas.
# El modelo final deberá predecir la probabilidad de lluvia del día de mañana a partir de las condiciones del día de hoy.
# El data set utilizado abarca datos desde noviembre de 2007 a junio del 2017
#


# 1. Importar la librería Pandas
import pandas as pd

# 2. Asignar la Raw URL a una variable
url_github = 'https://raw.githubusercontent.com/esprado1976/data_files/refs/heads/main/weatherAUS.csv'

# 3. Cargar el dataset directamente desde la URL usando pd.read_csv()
try:
    df = pd.read_csv(url_github)
    print("¡Dataset cargado exitosamente!")

# 4. Generar un dataset solo con los registros que tienen todos sus datos compeltos (sin NaN)
    df_limpio = df.dropna(how='any')

# 5. Mostrar las primeras 5 filas para verificar
    print("\nPrimeras 5 filas del DataFrame:")
    print(df_limpio.head())

# 6. Contabilizar el total de valores
    total_valores = df_limpio.size
    print(f"\nTotal de valores en el DataFrame: {total_valores}")

# 7. Determinar el rango temporal
    fecha_inicio = df_limpio['Date'].min()
    fecha_fin = df_limpio['Date'].max()
    print(f"Período de estudio: {fecha_inicio} a {fecha_fin}")

# 8. Determinar cantidad de locaciones
    print(df_limpio['Location'].value_counts())


except Exception as e:
    print(f"Ocurrió un error al cargar el dataset: {e}")
    print("Asegúrate de que la URL sea la 'Raw URL' de GitHub y que el archivo esté bien formado.")




```
